```python
df = pd.read_csv('/content/drive/MyDrive/Demo Datasets/Social_Network_Ads.csv')
df.head()
```

```python
df.info()
```

```python
df['Purchased'].value_counts()
```

```python
gender = pd.get_dummies(df['Gender'], drop_first=True)
df = pd.concat([df, gender], axis=1)
```

```python
df.head()
```

```python
df.drop(['Gender'],axis=1, inplace=True)
```

```python
x = df[['Age','EstimatedSalary','Male']]
y = df['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
```

```python
scaled_feature = scaler.transform(x)
scaled_feature
```

```python
df_feat = pd.DataFrame(scaled_feature, columns = x.columns)
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_feature, y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
confusion_matrix(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```
